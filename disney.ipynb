```python
#Importing Dependencies
import pandas as pd
from sqlalchemy import create_engine
import datetime as dt
from config import password
```

# Extracting Data

```python
#Importing the JSON file (The whole Raw data)
json_file = "Resources/disney_movies.json"
json_movies = pd.read_json(json_file)

#Listing all the columns in the json file
json_movies.head(10)
```

```python
#Extracting only the Columns needed for the Base Table
movies_df = json_movies[['title', 'Release date','Directed by', 'imdb_rating' ,'Running time',
                         'Budget (float)']].copy()
movies_df.head(20)
```

```python
dis_movies = "Resources/disney_movies.csv"
dis_movies = pd.read_csv(dis_movies)
dis_movies.head()
```

```python
characters = "Resources/disney-characters.csv"
characters_df = pd.read_csv(characters)
characters_df.head()
```

```python
VA = "Resources/disney-voice-actors.csv"
VA = pd.read_csv(VA)
VA.head(10)
```

# Cleaning the Data

```python
#Renaming the Column Names
movies_df = movies_df.rename(columns={'title': 'Title', 'Release date': 'Release_Date','Directed by': 'Director', 'Running time': 'Run Time(mins)',
                          'Budget (float)':'Budget (USD in Millions)', 'imdb_rating' :'IMDB Rating' } )
movies_df.head()
```

```python
#Formatting the Release Date column
movies_df['Release_Date' ] = movies_df['Release_Date'].str[0]
new_df  = movies_df["Release_Date"].str.split("(", n = 1, expand = True)
movies_df["Release_Date"] = new_df[0]
movies_df.head()
```

```python
#Formatting the movie_title Column in Characters_df to remove '\n' Characters
characters_df['movie_title']=  characters_df['movie_title'].str.replace('\r\n', '')
characters_df.head(20)
```

# Transforming Data

```python
movies_df
```

```python
movies_df.loc[movies_df['Title'] == "Alice in Wonderland"]
```

```python
dis_movies = dis_movies[["movie_title", "genre", "mpaa_rating"]]
dis_movies
```

```python
#Merging two dataframes to put together basic information of each movie
whole_df = movies_df.merge(dis_movies, left_on = 'Title', right_on = 'movie_title')

#Viewing the whole merged dataframe
whole_df
```

```python
#Viewing the columns of merged dataframe to grab the ones that are required
```

```python
whole_df = whole_df.drop_duplicates(subset=['Release_Date'])
whole_df = whole_df.reset_index(drop=True)
basic_df = whole_df[['Title', 'Release_Date', 'Director','Run Time(mins)','mpaa_rating', 'genre' ]]
basic_df = basic_df.sort_values(by=['Title'])
basic_df = basic_df[(basic_df['Release_Date'].str.len() > 9)]
basic_df = basic_df.reset_index(drop=True)
basic_df = basic_df.reset_index(drop=False)
basic_df = basic_df.rename(columns = {"index":"id", "Title":"title", "Release_Date":"release_date", "Director":"director",
                                      "Run Time(mins)":"run_time", 'mpaa_rating':"age_rating", "genre":"genre"})
basic_df
```

```python
#basic_df['length'] = basic_df['Release_Date'].str.len()
#basic_df.sort_values('length', ascending=True, inplace=True)
#basic_df
```

```python
ratings_df = json_movies[["title", "imdb_rating", "metascore", "rotten_tomatoes"]]
ratings_df = basic_df.merge(ratings_df, left_on = "title", right_on = "title")
ratings_df = ratings_df[["title", "release_date", "imdb_rating",'rotten_tomatoes']]
ratings_df = ratings_df.drop_duplicates(subset=['release_date'])
ratings_df = ratings_df.dropna()
ratings_df = ratings_df[["title", "release_date", "imdb_rating",'rotten_tomatoes']]
ratings_df = ratings_df.rename(columns = {"title": "title",
                                        "imdb_rating": "imdb",
                                          "rotten_tomatoes": "rotten_tomatoes"})
ratings_df = ratings_df.reset_index(drop=True)
ratings_df.head(20)
```

```python
cast_df = basic_df.merge(characters_df, left_on='title', right_on= 'movie_title')
cast_df = cast_df[['title', 'release_date_x', 'hero', 'villian']]
cast_df = cast_df.drop_duplicates(subset=['title'])
cast_df = cast_df.merge(VA, left_on = "title", right_on = "movie")
cast_df = cast_df[['title', 'release_date_x', 'hero', 'villian', 'character', 'voice-actor']]
cast_df = cast_df.rename(columns = {'release_date_x':'release_date', 'voice-actor': 'voice_actor', 'villian':'villain'})

cast_df
```

```python
crew_df = json_movies.merge(basic_df, left_on= 'title', right_on='title', how = 'inner')
crew_df = crew_df[['title', 'release_date', 'Music by', 'Cinematography','Edited by','Screenplay by']]
crew_df = crew_df.rename(columns = {'Music by':"composer",
                                      'Cinematography': 'cinematographer',
                                     'Edited by' :'editor',
                                     'Screenplay by': 'screenplay' })

crew_df = crew_df.drop_duplicates(subset=['release_date'])
crew_df.head(10)
```
